# Сборный проект. Компьютерные игры

**Заказчик** - интернет-магазин «Стримчик», который продаёт по всему миру компьютерные игры. 

**Данные, которые нужно изучить** - из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Данные до 2016г. Нужно планировать компанию на 2017 г.

**Бизнес-задача** - нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

**Описание данных:**

- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

**План действий:** изучить общую информацию о файле с данными, сделать предобработку данных (заполнить пропущенные данные, тип данных поменять, проверить на дубликаты и т.д.), сделать анализ на основе данных, проверить гипотезы, сделать общий вывод.

Для проверки рабочих гипотез особенно ценны столбцы: Platform, Year_of_Release, Genre, NA_sales, EU_sales, JP_sales, Other_sales, Critic_Score, User_Score, Rating 

## Шаг 2. Подготовка данных

```python
games.columns = games.columns.str.lower()
games.head()
```

```python
games.isna().sum()# узнаем суммарное количество пропусков, выявленных методом isnull() в таблице
```

```python
games['user_score'].value_counts()
```

```python
import numpy as np
games['user_score'] = games['user_score'].replace('tbd', np.nan, regex=True)
games['user_score'].value_counts()
```

```python
games['genre'].unique()#смотрим уникальные значения жанров
```

```python
games['rating'].value_counts()#смотрим значения рейтинга ESRB
```

```python
games.dropna(subset = ['year_of_release', 'genre', 'name'], inplace = True)#удаляем строки имеющие пропуски
games.isna().sum()
```

```python
games['year_of_release'] = games['year_of_release'].astype(int)
games['user_score'] = pd.to_numeric(games['user_score'])#меняем на числовой тип
games.dtypes
```

```python
games.duplicated().sum()#смотрим есть ли дубликаты
```

```python
games['sum_sales'] = games['na_sales'] + games['eu_sales'] + games['jp_sales'] + games['other_sales']#создаем новый столбец с общим доходом от кааждой игры
games.head(10)
```

### Вывод

- Привели название столбцов на нижний регистр

Поменяли тип данных:
- Столбец 'year_of_release' перевели на числовой тип int. Почему я не перевела на тип данных даты? Потому что, столбец содержит данные только о годе выхода игры и в последующем будет легче работать с данным в типе int, так как есть только одна цифра - год.
- Столбец 'user_score' переевли на числовой тип данных

Пропуски:
- **Пропуски в столбце 'critic_score', 'user_score', 'rating' оставила в исходном положении, чтобы не исказить данные по рекомендации ревьюера**
**На столбце 'user_score' есть строка со значением tbd, что означает "to be declared" – подробности могут быть определены, но еще не готовы к объявлению. Их перевела на np.nan по рекомендации ревьюера**

- Пропуски в столбце 'year_of_release'  удалила, количество пропусков 1.6% и их отсутствие не повлияют на результат.

Дубликаты не были обнаружены.

- Важны ли данные за все периоды?
Мы хоти планировать компанию на 2017г., поэтому нам не важны все годы, а чтобы выбрать важные нам годы дальше посмотрим, срок жизни платформ и в соответствии с этим сроком узнаем важные нам годы.

```python
games.groupby('platform')['sum_sales'].agg('sum').nlargest(20)#выбираем платформы с наибольшими суммарными продажами
```

```python
#выбираем платформы с наибольшими суммарными продажами, чтобы построить распределение по годам
selected_platform = games.query('platform ==("PS2", "X360", "PS3", "Wii", "DS", "PS", "PS4", "GBA", "PSP", "3DS", "XB", "PC", "XOne")')
selected_platform.platform.unique()
```

```python
import seaborn as sns#строим график
platform_life = selected_platform.pivot_table(index='year_of_release', columns='platform', values='sum_sales', aggfunc='sum')
sns.lineplot(data=platform_life, dashes=False)
plt.title('Срок жизни платформ по годам и прибыль')
plt.xlabel('Годы')
plt.ylabel('Продажи')
plt.show()
```

Из построенного графика можем сделать вывод срок жизни платформы в среднем - 6 лет. И теперь, чтобы сделать прогноз на 2017год будем смотреть на результаты 2011-2016годов - это будет для нас актуальным периодом. Дальше будем работать с данными в этих годах, так как они нам важны для выявления результата.
Теперь хоти узнать, какие платформы популярны в 2011-2016годах, для этого делаем срез данных по этим годам.

```python
games_act_period = games.query('year_of_release > 2010')
```

```python
d = sns.catplot(x='year_of_release', y='sum_sales', data=games_act_period, kind='bar', hue='platform', saturation=.5,ci=None, aspect=.6, palette="pastel", edgecolor=".6")
d.fig.set_size_inches(15,8)
d.fig.subplots_adjust(top=0.81,right=0.86) 
plt.title('Платформы имеющие успех за 2011-2016г.')
plt.xlabel('Годы')
plt.show()
```

Выбираем как потенциально прибыльными платформами в 2011-2016годах - Wii, DS, PS4, PS3, X360, PS2, 3DS, PSP, XOne, WiiU, PC, PSV. Дальше для этих платформ построим график «ящик с усами» по глобальным продажам игр на актуальный для нас период

```python
final_data = games.query('platform ==("PS2", "X360", "PS3", "Wii", "DS", "PSV", "PS4", "WiiU", "PSP", "3DS", "PC", "XOne") and year_of_release > 2010')
final_data.year_of_release.unique()
```

```python
#хотим узнать какие платформы лидируют по продажам, растут или падают, строим график "ящик с усами"

sns.boxplot(x='platform', y='sum_sales', palette='rainbow', data=final_data,  showfliers=False)
plt.title('Платформы с суммарными продажами')
plt.xlabel('Платформы')
plt.ylabel('Продажи')
plt.show()
```

Из графика «ящик с усами» видим, что больше всего дохода за 2011-2016год имеют платформы - PS3, X360, PS4, XOne, WiiU. А у платформ PSV, PSP, PS2 - доходы в эти года уже меньше, потому что уже срок жизни этих платформ заканчивается. 
Дальше хотим больше изучить 5 платформ которые имеют наибольший доход и узнать как влияют на продажи отзывы критиков и пользователей. Для этого построим диаграмму рассеяния и посчитаем корреляцию между этими значениями.

```python
games_ps3= final_data.query('platform == "PS3"') #выбираем данные только по платформе PS3
games_ps3.plot(x='sum_sales', y='critic_score', kind='hexbin', gridsize=5, figsize=(8, 6), sharex=False, grid=True) 

plt.title('Диаграмма рассеяния для платформы PS3: оценка критиков и продажи')
plt.xlabel('Продажи')
plt.ylabel('Оценка критиков')
plt.show()

games_ps3['sum_sales'].corr(games_ps3['critic_score'])#смотрим корреляцию между отзывом критиков и общим доходом
```

```python
games_ps3= final_data.query('platform == "PS3"') #выбираем данные только PS3
games_ps3.plot(x='sum_sales', y='user_score', kind='hexbin', gridsize=5, figsize=(8, 6), sharex=False, grid=True) 

plt.title('Диаграмма рассеяния для платформы PS3: оценка пользователей и продажи')
plt.xlabel('Продажи')
plt.ylabel('Оценка пользователей')
plt.show()

games_ps3['sum_sales'].corr(games_ps3['user_score'])#смотрим корреляцию между отзывом пользователей и общ
```

Из диаграммы рассеяния видно, что даже если по отзывам большие оценки - это не приносит больших доходов. Но при этом по пользовательским оценкам получившие низкие оценки как 0, у них доходов меньше. ## Шаг 5. Гипотезы

```python
final_data.dropna(subset = ['user_score'], inplace = True)#удаляем строки имеющие пропуски
final_data.isna().sum()
```

```python
from scipy.stats import levene#тест Левена
sample_xone = final_data.query('platform == "XOne"')['user_score']
sample_pc = final_data.query('platform == "PC"')['user_score']
stat, p = levene(sample_xone, sample_pc)
p
```

```python
[np.var(x, ddof=1) for x in [sample_xone, sample_pc]]#дисперсия
```

```python
from scipy import stats as st#гипотеза

alpha = .05 # критический уровень статистической значимости
    
results = st.ttest_ind(sample_xone, sample_pc, equal_var = False)
    
print('p-значение: ', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Нужно было проверить гипотезу - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые. Для этого я написала 
- Нулевая гипотеза: Средние пользовательские рейтинги платформ Xbox One и PC равны
- Алтернативная гипотеза: Средний пользовательский рейтинги платформы Xbox One больше (или меньше),чем средний пользовательский рейтинги платформы РС

1. Чтобы это решить, избавилась от значении nan, которые помешают результату.
2. Воспользовалась стат.тестом Левена, где задала 2 выборки. И получила p-value равный 0.019 . Если p-value Левена > уровня значимости, то equal_var= True. В ином случае равен False У меня уровень значимости равен 0.05, поэтому equal_var = False. Малое значение p предполагает, что совокупности не имеют одинаковых дисперсий.

```python
from scipy.stats import levene#тест Левена
sample_act= final_data.query('genre == "Action"')['user_score']
sample_sp= final_data.query('genre == "Sports"')['user_score']

stat, p = levene(sample_act, sample_sp)
p
```

```python
[np.var(x, ddof=1) for x in [sample_act, sample_sp]]#дисперсия
```

```python
from scipy import stats as st

alpha = .05 # критический уровень статистической значимости
    
results = st.ttest_ind(sample_act, sample_sp, equal_var = True)
    
print('p-значение: ', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Воспользовалась стат.тестом Левена, где задала 2 выборки. И получила p-value равный 1.98 . Если p-value Левена > уровня значимости, то equal_var= True. В ином случае равен False. У меня уровень значимости равен 0.05, поэтому equal_var = True. Малое значение p предполагает, что совокупности не имеют одинаковых дисперсий.

- Алтернативная гипотеза: Средний пользовательский рейтинги платформы Xbox One больше (или меньше),чем средний пользовательский рейтинги платформы РС

Сформулируем гипотезы:

Почему сформулировала гипотезы так?
- Во первых, нулевая гипотеза не является вопросом, на который мы отвечаем. Нулевая гипотеза всегда формулирует равенство чего-то. И здесь в первом вопросе нулевую гипотезу задала как равенство двух средних пользовательских рейтингов по платформе и во втором варианте по жанрам.
- Альтернативная может формулироваться как не-равно, больше, меньше в зависимости от проверяемой гипотезы. И здесь уже указала что средний пользовательский рейтинг одной генеральной совокупности больше(или меньше), чем от другой генеральной совокупности.
